```python
df["Item"] = df["Item"].fillna(method="ffill")
df["Price Per Unit"].fillna(df['Price Per Unit'].median(),inplace=True)
df["Quantity"].fillna(df['Quantity'].median(),inplace=True)
df["Total Spent"].fillna(df['Total Spent'].median(),inplace=True)
df["Discount Applied"] = df["Discount Applied"].fillna(method="ffill")
```

```python
df.info()
```

```python
df.drop(['Transaction ID',"Customer ID"], axis=1, inplace=True)
```

```python
col= df.select_dtypes(include="object").columns
from sklearn.preprocessing import LabelEncoder
leb=LabelEncoder()
for i in col:
    df[i]=leb.fit_transform(df[i])

bol_leb=LabelEncoder
```
